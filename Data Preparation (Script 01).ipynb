```python
import pandas as pd

purchases = pd.read_csv('/Users/charisoneyemi/Downloads/THESISFOLDER/amazon-purchases.csv')
surveys = pd.read_csv('/Users/charisoneyemi/Downloads/THESISFOLDER/survey.csv')
column_details = pd.read_csv('/Users/charisoneyemi/Downloads/THESISFOLDER/fields.csv')
```

```python
column_mapping = dict(zip(column_details["Unnamed: 0"], column_details["fields"]))
new_survey_data = surveys.rename(columns = column_mapping)
```

```python
"""Basically changing the column names in the surveys csv to the ones in column_details"""
```

```python
"""To combine the transaction data (purchases) with the customer demographic data (surveys) so that each transaction now includes who made the purchase and their demographic details."""
```

```python
merged_dataset = purchases.merge( #left
    new_survey_data, #right
    how = "left", #keep all purchases ie our left table even if demographics (right) are missing (it will just be NaN)
    left_on = "Survey ResponseID",
    right_on = "Response ID" 
)
```

```python
merged_dataset.head(2)
```

```python
merged_dataset['Order Date'] = pd.to_datetime(merged_dataset['Order Date'])
#Here I'm converting the order date column to the date time format
```

```python
print(merged_dataset.isna().sum())  # to see which fields are missing
```

```python
# Filling product-level info since that’s useful for analysis
merged_dataset['Category'] = merged_dataset['Category'].fillna('UNKNOWN')
merged_dataset['Title'] = merged_dataset['Title'].fillna('UNKNOWN')
merged_dataset['Shipping Address State'] = merged_dataset['Shipping Address State'].fillna('UNKNOWN')
merged_dataset['ASIN/ISBN (Product Code)'] = merged_dataset['ASIN/ISBN (Product Code)'].fillna('UNKNOWN')
```

```python
merged_dataset = merged_dataset.rename(columns={
    'What is your age group?': 'Age Group',
    'Are you of Spanish, Hispanic, or Latino origin?': 'Hispanic Origin',
    'Choose one or more races that you consider yourself to be': 'Race',
    'What is the highest level of education you have completed?': 'Education',
    'What was your total household income before taxes during the past 12 months?': 'Household Income',
    'How do you describe yourself?': 'Gender Identity',
    'Which best describes your sexual orientation?': 'Sexual Orientation',
    '50 States, D.C. and Puerto Rico': 'State Reported',
    'How many people do you share your Amazon account with? i.e. how many people log in and make orders using your account?': 'Account Shared With',
    'How many people are in your "household"?': 'Household Size',
    'How often do you (+ anyone you share your account with) order deliveries from Amazon?': 'Amazon Order Frequency',

    # Health & behavior
    'Are any of the following the case for:\n\nYou or someone in your household or someone you share your Amazon account with? - Smoke cigarettes regularly?': 'Smoke Cigarettes',
    'Are any of the following the case for:\n\nYou or someone in your household or someone you share your Amazon account with? - Smoke marijuana regularly?': 'Smoke Marijuana',
    'Are any of the following the case for:\n\nYou or someone in your household or someone you share your Amazon account with? - Drink alcohol regularly?': 'Drink Alcohol',
    'Are any of the following the case for:\n\nYou or someone in your household or someone you share your Amazon account with? - Have diabetes?': 'Diabetes',
    'Are any of the following the case for:\n\nYou or someone in your household or someone you share your Amazon account with? - Use a wheelchair?': 'Wheelchair',

    # Life change event
    'In 2021 did you, or someone you share your Amazon account with, experience any of the following life changes?\n\nPlease mark all that apply.': 'Life Changes 2021',

    # Opinions
    'Do you think Amazon should be able to sell YOUR purchase data to other companies?': 'Opinion: Amazon Sell Data',
    'Do you think companies should be able to sell consumer purchase data to other companies?': 'Opinion: Companies Sell Data',
    'Big companies currently collect and sell consumer purchase data. Do you think that small businesses should be able to access this data for free in order to help them compete with the big companies?': 'Opinion: SMB Data Access',
    'Do you think the U.S. Census Bureau should use purchase data to supplement their existing surveys?': 'Opinion: Census Use Data',
    'Do you think researchers should be able to use purchase data to understand societal changes (e.g. due to COVID-19)?': 'Opinion: Research Use Data',
})
```

```python
merged_dataset.head(4)
```

```python
merged_dataset['Total_Spent'] = merged_dataset['Purchase Price Per Unit'] *  merged_dataset['Quantity']
```

```python
merged_dataset.head(1)
```

```python
# Grouping by customer ID and aggregating purchase behavior so we can have one row per customer with summary stats.
```

```python
customer_df = merged_dataset.groupby('Response ID').agg({
    'Order Date': ['min', 'max', 'nunique'],        
    'Quantity': 'sum',                              
    'Total_Spent': 'sum',                           
    'Title': 'nunique',                             
    'Category': 'nunique'                           
})
```

```python
customer_df.head()
```

The aggregation gives column names like


###### Order Date         min
###### Order Date         max
###### Order Date     nunique
###### Quantity           sum
###### Total_Spent        sum
###### Title           nunique
###### Category        nunique


This format is not easy to work with in code because you’d need to access columns like:

customer_df["Order Date"]["min"]


```python
customer_df.columns = [
    'First_Purchase', 'Last_Purchase', 'Purchase_Days',
    'Total_Quantity', 'Total_Spent', 'Unique_Products', 'Unique_Categories'
]
```

```python
customer_df.head()
```

```python
reference_date = merged_dataset['Order Date'].max()
# By Recency I mean how long before the last available purchase did this customer last buy something? 
#I'm going to be using the most recent order in the dataset

```

```python
customer_df['Recency_Days'] = (reference_date - customer_df['Last_Purchase']).dt.days
```

```python
customer_df['Frequency'] = customer_df['Purchase_Days']
#Just renaming Purchase_Days to Frequency for clarity
```

```python
customer_df['Avg_Order_Value'] = customer_df['Total_Spent'] / customer_df['Frequency']
# AOV is key to understanding customer value; higher spend per visit means higher value
```

```python
customer_df['Avg_Quantity_Per_Order'] = customer_df['Total_Quantity'] / customer_df['Frequency']
#This indicates if they buy in bulk or just one or two items per visit.
```

```python
customer_df = customer_df.reset_index()
#After grouping, Response ID became the index so I want to move Response ID back to being a column
```

```python
demographics = merged_dataset.drop_duplicates(subset='Response ID')
```


From the big merged dataset, this keeps only the first row for each unique user (Response ID)
Because each person may have made multiple purchases (multiple rows), but we only want one row per person for survey data

```python
survey_cols = [
    'Response ID', 'Age Group', 'Hispanic Origin', 'Race', 'Education',
    'Household Income', 'Gender Identity', 'Sexual Orientation', 'State Reported',
    'Account Shared With', 'Household Size', 'Amazon Order Frequency',
    'Smoke Cigarettes', 'Smoke Marijuana', 'Drink Alcohol', 'Diabetes', 'Wheelchair',
    'Life Changes 2021', 'Opinion: Amazon Sell Data', 'Opinion: Companies Sell Data',
    'Opinion: SMB Data Access', 'Opinion: Census Use Data', 'Opinion: Research Use Data'
]
#The relevant columns I need, so they can be included in the final dataset
```

```python
demographics = demographics[survey_cols]
# This keeps only the columns listed in survey_cols in the demographics dataframe
```

```python
final_dataset_DS = customer_df.merge(demographics, on="Response ID", how="left")
```
